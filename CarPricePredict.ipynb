### Creating Model

```python
#Splitting input and output columns
X = cars.drop(columns='Price')
y = cars['Price']
```

```python
X #input
```

```python
y #ouput
```

#### Performing Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
#Since it a prediction project on numerical data(price), we have to use Regression. 
#So, we using Linear regression
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
```

```python
Name	Location	Year	Kilometers_Driven	Fuel_Type	Transmission	Owner_Type	Mileage	Engine	Power	Seats
0	Maruti Wagon	Mumbai	2010	
```

```python
cars.head(5)
```

```python
ohe = OneHotEncoder()
ohe.fit(X_train[['Name','Location','Fuel_Type','Transmission','Owner_Type','Type']])
```

### Creating a column transformer to transform categorical columns

```python
from sklearn.compose import make_column_transformer
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_,handle_unknown="ignore"),['Name','Location','Fuel_Type','Transmission','Owner_Type','Type']),
                                    remainder='passthrough')
```

### Linear Regression Model

```python
lr = LinearRegression()
```

### Making a pipeline

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(column_trans,lr)
```

### Fitting the model

```python
pipe.fit(X_train,y_train)
```

### Predict the outputs

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

### Checking R2 Score

```python

r2_score(y_test,y_pred)
```

### Finding the model with a random state of TrainTestSplit where the model was found to give maximum r2_score

- Creating a model in a loop and checking the maximum r2 score

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```
